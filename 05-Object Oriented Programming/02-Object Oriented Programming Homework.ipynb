#### Problem 1
Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.

```python
from math import sqrt
class Line:

    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2

    #
    # distance formula
    # d=sqrt((x2-x1)^2 + (y2-y1)^2)
    def distance(self):
        return sqrt((self.coor2[0] - self.coor1[0])**2  +  (self.coor2[1] - self.coor1[1])**2)
    #
    # slope formula
    # m=(y2-y1)/(x2-x1)
    def slope(self):
        return  (self.coor2[1] - self.coor1[1]) / (self.coor2[0] - self.coor1[0])

      #9.433981132056603
```

```python
# EXAMPLE OUTPUT

coordinate1 = (3,2)   # (x1,y1)
coordinate2 = (8,10)   # (x2,y2)

li = Line(coordinate1,coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

#### Problem 2

Fill in the class

```python
class Cylinder:
    # class object attributes
    # same for all the instances of the class
    pi = 3.14

    def __init__(self,height=1,radius=1):
        self.height = height
        self.radius = radius

    #
    #  the volume of a cylinder is V=pi*(r**2)h
    def volume(self):
        return Cylinder.pi * (self.radius**2) * self.height
    #
    #  the surface area of a cylinder is V=2*pi*r(r+h)
    def surface_area(self):
        return 2 * Cylinder.pi * self.radius * (self.radius + self.height)
```

```python
# EXAMPLE OUTPUT
c = Cylinder(2,3)
```

```python
c.volume()
```

```python
c.surface_area()
```
