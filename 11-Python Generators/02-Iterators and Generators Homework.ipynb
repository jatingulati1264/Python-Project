# Iterators and Generators Homework

### Problem 1

Create a generator that generates the squares of numbers up to some number N.

```python
def gensquares(N):
  for x in range(N):
    yield x**2
```

```python
for x in gensquares(10):
    print(x)
```

### Problem 2

Create a generator that yields "n" random numbers between a low and high number (that are inputs). <br>Note: Use the random library. For example:

```python
import random

random.randint(1,10)
```

```python
def rand_num(low,high,n):
  for x in range(n):
    yield random.randint(low,high)
```

```python
for num in rand_num(1,10,12):
    print(num)
```

### Problem 4
Explain a use case for a generator using a yield statement where you would not want to use a normal function with a return statement.<br>

We can use generator wherever we want to get a sequence of values with memory efficency eg-> fibonacci generator

In the normal functions whereever we are getting the larger amount of memory there we will use the generators.
    


```python
def gen_fibo(n):
  a = 0
  b = 1
  for x in range(n):
    yield a
    a,b = b, a+b

for number in gen_fibo(10):
  print(number)
```
